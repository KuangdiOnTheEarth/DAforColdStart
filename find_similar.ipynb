```python
from gensim.models.word2vec import Word2Vec
```

```python
seq_list = []
dataset = 'ml-1m'
f = open('data/%s/formatted/sequences.txt' % dataset, 'r')
for line in f:
    line_seg = line.rstrip().split(' ')
    seq_list.append(line_seg[1:])

model = Word2Vec(seq_list, window=5, min_count=1, workers=4)
```

```python
import codecs
movies = {}
cs_items = []
f = codecs.open('data/%s/formatted/movies.txt' % dataset, 'r', encoding='iso-8859-15')
for line in f:
    line_seg = line.rstrip().split('::')
    id = line_seg[0]
    movies[id] = line_seg[1]
f = open('data/%s/splits/cs_item_list.txt' % dataset, 'r')
for line in f:
    cs_items.append(line.rstrip())
print(cs_items)
```

```python
id = '3344'
print(movies[id])
similar_list = model.wv.most_similar(id)
for e in similar_list:
    print("%s %s" %(str(e), movies[e[0]] ))
```

```python
id = '3431'
# id = '3408'
print(movies[id])
similar_list = model.wv.most_similar(id, topn=100)
rank = 0
for e in similar_list:
    rank += 1
    if e[0] not in cs_items:
        print("%d %s %s" %(rank, str(e), movies[e[0]] ))
```

```python
output_f = open('data/%s/splits/similar_items.txt' % dataset, 'w')
for cs_item in cs_items:
    synonyms_str = ''
    count = 0
    similar_list = model.wv.most_similar(cs_item, topn=500)
    for iid_score_pair in similar_list:
        iid = iid_score_pair[0]
        if iid not in cs_items:
            synonyms_str += (' ' + iid)
            count += 1
        if count >= 3: break
    if count < 3: print("%s: only %d similar items found" % (cs_item, count))
    output_f.write("%s%s\n" % (cs_item, synonyms_str))
output_f.close()
print('Done')
```

```python
model.save("word2vec/word2vec.model")
```
